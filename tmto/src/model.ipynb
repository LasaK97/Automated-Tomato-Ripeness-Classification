```python
from ultralytics import YOLO
import time
```

```python
# YOLOv8 Model
model = YOLO("yolov8m.pt")
```

```python
data = """
names:
  - b_fully_ripened
  - b_half_ripened
  - b_green
  - l_fully_ripened
  - l_half_ripened
  - l_green
nc: 6
path: ../
train: ../data/train/images
val: ../data/val/images
"""

# save the text
file_path = "../data/dataset.yaml"

# write the text
with open(file_path, "w") as file:
    file.write(data)

print("ready")
```

```python
# Model Trianing
st = time.time()
results = model.train(data='../data/dataset.yaml', epochs=100, imgsz=640, model="yolov8m.pt", batch=16) 
et = time.time()

duration = et - st
hours = duration // 3600
minutes = (duration % 3600) // 60
seconds = duration % 60

print(f"\nDuration: {int(hours)} hours, {int(minutes)} minutes, and {seconds:.2f} seconds")
```

```python
from PIL import Image
from IPython.display import display

def display_images(image_paths, desired_size=(700, 600)):
    for path in image_paths:
        image = Image.open(path)
        image.thumbnail(desired_size)
        display(image)

# List of image paths
image_paths = ["../src/runs/detect/train2/results.png",
               "../src/runs/detect/train2/confusion_matrix.png",
               "../src/runs/detect/train2/labels.jpg"]

# Display all images with the specified size
display_images(image_paths)
```
